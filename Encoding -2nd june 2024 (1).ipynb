## Encoding
- converting categorical data into continous data

## 1) Label Encoding


- usually we apply label encoding to the data of ordinal type
- ordinal categorical data
- where data follows particular sequence
- days of the week, -sun, mon,tues, wed,thurs, - 0- sunday, 1 -monday, 2- tuesday, 3- wednesday?

## 2) One hot encoding
- The other method of encoding is one hot enocoding, where your data does not follow any sequence.
- e.g Gender male-0, female-1

## 3) Ordinal Encoding

```python
#step1 - import the package
from sklearn.preprocessing import LabelEncoder
```

```python
#step2- object creation
nirmal = LabelEncoder()
```

```python
#step3 - fit_transform
df['sex_label_encoding'] = nirmal.fit_transform(df['sex'])
```

## Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```
